```python
import math
def transform(CPU_list, base = 10):
    max_val = max(CPU_list)
    min_val = min(CPU_list)
    min_val_log = math.log(min_val,base)
    max_val_log = math.log(max_val,base) 

    res = []
    for cpu in CPU_list:

        res.append(math.log(cpu, base))
    return res
```

```python
RAM_er_ID = []
RAM_transform = []
RAM_tmp = []
for i,r in enumerate(ram):
    r = r.lower()
    # Define the pattern using regular expression
    pattern = re.compile(r'(\d+)\s*(gb|mb|tb)')

    # Find all matches in the input string
    matches = pattern.findall(r)
    
    # Print the matches
    print("Matches:", matches)
    if len(matches) < 1:
        storage_er_ID.append(i)
    else:
        if len(matches[0]) < 2:
            storage_er_ID.append(i)
        else:
            value = int(matches[0][0])
            type = matches[0][1]
            if type == "tb":
                value = value*1024
            elif type == "mb":
                value == value / 1024
                
    RAM_tmp.append(value)
RAM_transform = transform(RAM_tmp)
```

```python
monitor_er_ID = []
monitor_transform = []
mon_tmp = []
for i,m in enumerate(monitor):
    m = m.lower()
    # Define the pattern using regular expression
    pattern = re.compile(r'(\d+)\s*(GB|MB|TB)?\s*(?:inches|inch|")?')

    # Find all matches in the string
    matches = pattern.finditer(m)

    extracted_digits = []
    for match in matches:
        # Extract the digits from each match
        digits = match.group(1)
        extracted_digits.append(digits)
    if len(extracted_digits) == 0:
        print(m, matches)
    if len(extracted_digits) >=  2:
        float_coms = [extracted_digits[0], extracted_digits[1]]
        combined_number = float('.'.join(float_coms))
    else:
        combined_number = int(extracted_digits[0])
    mon_tmp.append(combined_number)

monitor_transform = transform(mon_tmp)
print(len(monitor_transform))
print(len(monitor_er_ID))
```

```python
import re
resolution_er_ID = []
resolution_transform = []
res_tmp = []
cnt = 0

digits = "0123456789"
for i,res in enumerate(resolution):
    pattern = re.compile(r'\d{3,4}x\d{3,4}')

    res = res.replace(" ","")
    matches = pattern.findall(res)
    if len(matches) != 1:
        cnt += 1

        resolution_er_ID.append(i)
    else:
        tmp = matches[0]
        w,h = tmp.split("x")

        width = int(w)
        height = int(h)
        tmp = width*height
        res_tmp.append(tmp)
        

resolution_transform = transform(res_tmp)
    


```
