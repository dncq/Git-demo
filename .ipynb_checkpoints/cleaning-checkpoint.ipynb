```python
def brand_duplicate(brand_name):
    name = brand_name.lower()
    if ('dell' in name) or ('envy' in name) or ('alienware' in name):
        return 'Dell'
    elif 'asus' in name:
        return 'Asus'
    elif 'acer' in name:
        return 'Acer'
    elif 'lenovo' in name:
        return 'Lenovo'
    elif 'toughbook' in name:
        return 'Panasonic'
    elif 'hp' in name:
        return 'HP'
    elif 'xpg' in name:
        return 'Adata'
    elif 'msi' in name:
        return 'MSI'
    elif 'samsung' in name:
        return 'Samsung'
    else:
        return brand_name
```

```python
def monitor_change(size):
    try:
        low_size = size.lower()
        if 'inches' in low_size:
            return '%.1f'%float(low_size.split()[0])
        elif 'centimeters' in low_size:
            return '%.1f'%(float(low_size.split()[0]) / 2.54)
        else:
            return '%.1f'%float(size)
    except:
        pass
        
```

```python
def weight_change(weight):
    try:
        split_weight = weight.lower().split()
        if 'pounds' in split_weight:
            return '%.2f' % (float(split_weight[0]) / 2.2)
        elif split_weight[1] == 'grams':
            return '%.2f' % (float(split_weight[0]) / 1000)
        elif 'ounces' in split_weight:
            return weight
        else:
            return '%.2f' % (float(split_weight[0]))
    except:
        pass
```

# Clean resolution

```python
df['Resolution'].unique()
```

```python
def find_resolution_patterns(resolution):
    try:
        resolution_name = resolution.replace('_','').replace("*",'x').replace("X",'x')
        pattern = r'\b\d+\s*[xX]\s*\d+\b'
        matches = re.findall(pattern, resolution_name)
        if matches:
            return list(map(int,matches[0].split('x')))
        else:
            return None
    except:
        return None
```

```python
widths = []
heights = []

for res in df['Resolution']:
    found = find_resolution_patterns(res)
    if found:
        widths.append(found[0])
        heights.append(found[1])
    else:
        widths.append(found)
        heights.append(found)
```

# Clean Storage

```python
df.Storage.unique()
```

```python
df['Storage'].value_counts()
```

```python
def get_storage_amount(st):
    try:
        st_lst = st.lower().split()
        if len(st_lst) <= 1:
            return float(st_lst[0])
        else:
            if 'tb' in st_lst:
                return float(float(st_lst[0]) * 1024 * 1024 )
            elif 'gb' in st_lst:
                return float(float(st_lst[0]) * 1024)
            elif 'mb' in st_lst:
                return float(st_lst[0])
            else:
                return float(float(st_lst[0]) * 1024)
    except:
        return None
    
def get_storage_type(st):
    try:
        st_low = st.lower()
        if (',' in st_low) or ('hybrid' in st_low):
            return 'Hybrid'
        else:
            if ('ssd' in st_low) or ('solid' in st_low):
                return 'SSD'
            elif 'hdd' in st_low:
                return 'HDD'
            elif ('emmc' in st_low) or ('embedded' in st_low):
                return 'Emmc'
            elif 'ata' in st_low:
                return 'SATA'
            elif 'flash' in st_low:
                return 'FLash'
            else:
                return 'HDD'
    except:
        return None
```

```python
storage_name = '500 GB Serial ATA-300'
print(get_storage_type(storage_name))
```

```python
storage_amount = [get_storage_amount(st) for st in df['Storage']]
storage_type = [get_storage_type(st) for st in df['Storage']]

print(len(storage_amount))
print(len(storage_type))
```

```python
print(storage_type)
```

```python
df['Storage Amount'] = storage_amount
df['Storage Type'] = storage_type
```

# Clean OS

```python
df[df.OS == 'Windows']
```

```python
df['OS'].unique()
```

```python
def extract_windows_patterns(input_string):
    pattern = r'(?:Windows|Win)\s\d+|\b(?:Windows|Win)\s\w+\b'
    matches = re.findall(pattern, input_string)
    return matches

input_string = 'Windows 10'
output = extract_windows_patterns(input_string)
print(output)
```

```python
def clean_os(os):
    try:
        os_low = os.lower().replace("\\"," ")
        if 'win' in os_low:
            if os_low == 'windows':
                augmented = ['7','8.1','10','11','Vista']
                return 'Windows ' +  random.choice(augmented)
            elif 'windowws' in os_low:
                return 'Windows 10'
            else:
                d = extract_windows_patterns(os)[0]
                if 'Windows' in d:
                    return d
                else:
                    return 'Windows' + ' ' + d.split()[1]
        elif 'mac' in os_low:
            return "MacOS"
        elif 'chrome' in os_low:
            return "ChromeOS"
        elif os_low == 'w':
            return 'No OS'
        else:
            return os
    except:
        return "No OS"
```

```python
new_os = [clean_os(os) for os in df['OS']]
len(new_os)
```

```python
df['New OS'] = new_os
```

```python
df['New OS'].value_counts()
```

# Levenshtein Distance

```python
from Levenshtein import ratio
"""
def LCS(s, t):
    n = len(s)
    m = len(t)
    s=' '+s
    t=' '+t

    f = [[0 for i in range(m + 1)] for j in range(n+1)]
     
    for i in range(1,n + 1):
        for j in range(1,m + 1):
            if(s[i - 1] == t[j - 1]):
                f[i][j] = f[i - 1][j - 1] + 1
            else:
                f[i][j] = max(f[i-1][j], f[i][j-1])
    return (f[n][m]+n+m)/(n+m)
"""
def find_relative_string(s, slist):
    found=[-1, -1]
    for i in range(len(slist)):
        acc=ratio(s, slist[i])
        if found[0]<acc:
            found[0]=acc
            found[1]=i
    return found

def transform_cpu(row):
    cpu = row['CPU']
    if pd.notna(cpu):
        cpu=cpu.lower()
        acc, pos=find_relative_string(cpu, cpu_name_list)
        if acc>0.5:
            row["CPU Name"] = cpu_df['CPU Name'][pos]
            row["CPU Mark"] = cpu_df['CPU Rank'][pos]
        else:
            row["CPU Name"] = np.nan
            row["CPU Mark"] = np.nan
    return row

def transform_gpu(row):
    gpu = row['GPU']
    if pd.notna(gpu):
        gpu=gpu.lower()
        acc, pos=find_relative_string(gpu, gpu_name_list)
        if acc>0.5:
            row["GPU Name"] = gpu_df['GPU Name'][pos]
            row["GPU Mark"] = gpu_df['GPU Rank'][pos]
        else:
            row["GPU Name"] = np.nan
            row["GPU Mark"] = np.nan
    return row
```

# Fuzzy String Matching for Amazon data

```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process

def mapping(s, slist):
    s2 = s.lower()
    found=[0, -1]
    for i in range(len(slist)):
        if s2 in slist[i]:
            found[0] = 100
            found[1] = i
            break
        acc=sum([fuzz.ratio(slist[i], s2),fuzz.partial_ratio(slist[i], s2),fuzz.token_sort_ratio(slist[i], s2),fuzz.token_set_ratio(slist[i], s2)])/4
#         acc=max(fuzz.ratio(slist[i], s2),fuzz.partial_ratio(slist[i], s2),fuzz.token_sort_ratio(slist[i], s2),fuzz.token_set_ratio(slist[i], s2))
        if found[0]<acc:
            found[0]=acc
            found[1]=i
    return found

def transform_cpu(row):
    cpu = row['CPU']
    if cpu:
        cpu=cpu.lower()
        acc, pos= mapping(cpu, cpu_name_list)
        if acc>60:
            row["CPU Name"] = cpu_df['CPU Name'][pos]
            row["CPU Mark"] = cpu_df['CPU Rank'][pos]
        else:
            row["CPU Name"] = np.nan
            row["CPU Mark"] = np.nan
    return row

def transform_gpu(row):
    gpu = row['GPU']
    if gpu:
        gpu=gpu.lower()
        acc, pos=mapping(gpu, gpu_name_list)
        if acc>60:
            row["GPU Name"] = gpu_df['GPU Name'][pos]
            row["GPU Mark"] = gpu_df['GPU Rank'][pos]
        else:
            row["GPU Name"] = np.nan
            row["GPU Mark"] = np.nan
    return row
```
