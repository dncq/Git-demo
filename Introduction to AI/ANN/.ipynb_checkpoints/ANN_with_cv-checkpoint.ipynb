```python
from emnist import list_datasets
import matplotlib as mpl
import matplotlib.pyplot as plt
import keras
import tensorflow as tf
from emnist import extract_training_samples, extract_test_samples
from keras.layers import Dense # Dense layers are "fully connected" layers
from keras.models import Sequential # Documentation: https://keras.io/models/sequential/
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, f1_score, precision_score, confusion_matrix, plot_confusion_matrix
from yellowbrick.model_selection import CVScores
import numpy as np
import itertools

#Balanced : 47 classes , each has 2400 training samples (112800) and 400 test samples (18800)
test_data, test_label =  extract_test_samples("balanced")
images, labels = extract_training_samples("balanced")

print('Images:',images.shape)
print('Labels:',labels.shape)
print('Test_data:',test_data.shape)
print('Test_labels:',test_label.shape)

my_data = images
my_label = labels

# Parse numbers as floats
my_data = my_data.astype('float32')
test_data = test_data.astype('float32')
# Normalize data
my_data = my_data / 255
test_data = test_data / 255
# Merge data and labels
x_data = np.concatenate((my_data,test_data))
y_label = np.concatenate((my_label,test_label))

print('x_data:',x_data.shape)
print('y_label:',y_label.shape)

# Split data
x_train,x_test,y_train, y_test = train_test_split(x_data,y_label,test_size = 7/25, random_state = 364,stratify = y_label)
print('x_train:',x_train.shape)
print('x_test:',x_test.shape)
print('y_train:',y_train.shape)
print('y_test:',y_test.shape)

# Flatten the images
image_vector_size = 28*28
image_size = 784
x_train = x_train.reshape(x_train.shape[0], image_vector_size)
x_test = x_test.reshape(x_test.shape[0], image_vector_size)
# test_data = test_data.reshape(test_data.shape[0], image_vector_size)
x_data = x_data.reshape(x_data.shape[0], image_size)
```

```python
def my_metrics(y_true, y_pred):
    """ Calculate accuracy, precision, and f1 score of the model's prediction with respect to true labels.

        Args:
            y_true: list/array
                All true class labels
            y_pred: list/array
                All predicted class labels

        Returns:
            accuracy: float
                Accuracy measure of the model
            precision: float
                Precision measure of the model
            f1_Score: float
                F1-score measure of the model

    """
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average='weighted')
    f1_Score = f1_score(y_true, y_pred, average='weighted')
    print("Accuracy  : {}".format(accuracy))
    print("Precision : {}".format(precision))
    print("f1Score : {}".format(f1_Score))
    cm = confusion_matrix(y_true, y_pred)
    print('Confusion matrix:')
    print(cm)
    return accuracy, precision, f1_Score, cm
```

```python
import matplotlib.pyplot as plt 

def get_model():
    model = Sequential()
    
    # The input layer requires the special input_shape parameter which should match
    # the shape of our training data.
    model.add(Dense(units=256, activation='relu', input_shape=(image_size,)))
    model.add(Dense(units=256, activation='elu'))
    model.add(Dense(units=47, activation ="softmax"))
    model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=['accuracy'])
    
    return model

model = get_model()

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting normalize=True.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1, keepdims = True)
    f = plt.figure()
    f.set_figwidth(100)
    f.set_figheight(100)

    plt.tight_layout()
    plt.title(title, fontsize = 80)
    plt.ylabel('True label', fontsize = 72)
    plt.xlabel('\nPredicted label', fontsize = 72)
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation = 45, fontsize = 40)
    plt.yticks(tick_marks, classes, fontsize = 40)
    plt.colorbar()
    
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),fontsize = 32, horizontalalignment="center",color="white" if cm[i, j] > thresh else "black")


class_names = [str(i) for i in range(47)]


```

```python
from sklearn import metrics
import matplotlib.pyplot as plt


num_classes = 47
# K-fold cross validation
acc_per_fold = []
loss_per_fold = []
f1_scores= []
folds = []
kf = KFold(n_splits = 5,shuffle = True)
fold_no = 1
for train, test in kf.split(x_data,y_label):
    print('------------------------------------------------------------------------')
    print(f'Training for fold {fold_no} ...')
    print(f'Train on {x_data[train].shape[0]} samples; Validate on {x_data[test].shape[0]} samples')
    # Fit the model
    history = model.fit(x_data[train], y_label[train],batch_size=32, epochs=15, verbose=True, validation_split=.1)
    # Generate generalization metrics
    scores = model.evaluate(x_data[test], y_label[test], verbose=False)
    print(f'Score for fold {fold_no}: {model.metrics_names[0]} of {scores[0]}; {model.metrics_names[1]} of {scores[1]*100}%')
    acc_per_fold.append(scores[1] * 100) # acc in percentage
    loss_per_fold.append(scores[0])
    # ======== Report =========================
    predict_x = model.predict(x_data[test]) 
    classes_x = np.argmax(predict_x,axis=1)
    print('========================REPORT==============================')
    print(classification_report(y_label[test],classes_x))
    print()
    val_acc, val_prec, val_f1_score, cm = my_metrics(y_label[test], classes_x)
    f1_scores.append(val_f1_score)
    # Plot the confusion matrix
#     plt.figure()
#     plot_confusion_matrix(cm,classes = class_names,title='Confusion Matrix\n\n')
#     plt.show()
    
#     plt.savefig(f'cnf_matrix_for_fold_no_{fold_no}.png')
    # Increase fold number
    folds.append(fold_no)
    fold_no = fold_no + 1
    
    
# == Provide average scores ==
print('SUMMARY:')
print('------------------------------------------------------------------------')
print('Score per fold:')
for i in range(0, len(acc_per_fold)):
  print('------------------------------------------------------------------------')
  print(f'> Fold {i+1} - Loss: {loss_per_fold[i]} - Accuracy: {acc_per_fold[i]}%')
print('------------------------------------------------------------------------')
print('Average scores for all folds:')
print(f'> Accuracy: {np.mean(acc_per_fold)} (+- {np.std(acc_per_fold)})')
print(f'> Loss: {np.mean(loss_per_fold)}')
print('------------------------------------------------------------------------')


# ===================Visualization=================================
plt.bar(folds, acc_per_fold,width = 0.3)
plt.title('Accuracy for each fold\n', fontsize = 18)
plt.axhline(np.mean(acc_per_fold),color = None,label="Mean score={:.3f}".format(np.mean(acc_per_fold)),linestyle="--",linewidth= 1)
plt.xlabel('Training Instances', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)
plt.grid(True)

# plt.bar(folds, f1_scores,width = 0.3)
# plt.title('f1_score for each fold\n', fontsize = 18)
# plt.axhline(np.mean(acc_per_fold),color = None,label="Mean score={:.3f}".format(np.mean(acc_per_fold)),linestyle="--",linewidth= 1)
# plt.xlabel('Training Instances', fontsize=14)
# plt.ylabel('Scores', fontsize=14)
# plt.grid(True)

plt.show()
# plt.plot(history.history['accuracy'])
# plt.plot(history.history['val_accuracy'])
# acc = history.history['val_accuracy']
# print(acc)
# plt.title('model accuracy

# plt.ylabel('accuracy')
# plt.xlabel('epoch')
# plt.legend(['training', 'validation'], loc='best')
# plt.show()


# print(f'Test loss: {loss:.3}')
```
